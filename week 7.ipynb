```python
# Program to find squares using map()

numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x * x, numbers))

print("Original List:", numbers)
print("Squares List:", squares)
```

```python
# Program to filter even numbers

numbers = [10, 15, 20, 25, 30, 35, 40]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print("Original List:", numbers)
print("Even Numbers:", even_numbers)
```

```python
# Program to find sum using reduce()

from functools import reduce

numbers = [1, 2, 3, 4, 5]

sum_result = reduce(lambda x, y: x + y, numbers)

print("List Elements:", numbers)
print("Sum of Elements:", sum_result)
```

```python
# Program demonstrating higher-order function

def multiplier(n):
    return lambda x: x * n

double = multiplier(2)
triple = multiplier(3)

print("Double of 5:", double(5))
print("Triple of 5:", triple(5))
```

```python
# Program to sort tuples using lambda

students = [
    ("Anu", 85),
    ("Rahul", 72),
    ("Priya", 90),
    ("Kiran", 78)
]

sorted_students = sorted(students, key=lambda x: x[1])

print("Students Sorted by Marks:")
for student in sorted_students:
    print(student)
```

```python
# Program to find maximum number using reduce()

from functools import reduce

numbers = [12, 45, 67, 23, 89, 34]

maximum = reduce(lambda x, y: x if x > y else y, numbers)

print("List Elements:", numbers)
print("Maximum Element:", maximum)
```

```python
from functools import reduce

# Employee salaries
salaries = [45000, 52000, 61000, 48000, 75000, 39000]

# Increase salary by 10%
increased_salary = list(map(lambda x: x * 1.10, salaries))

# Employees earning above 50,000
high_salary = list(filter(lambda x: x > 50000, salaries))

# Total salary expenditure
total_expenditure = reduce(lambda x, y: x + y, salaries)

print("Original Salaries:", salaries)
print("Salaries after 10% hike:", increased_salary)
print("Employees earning above 50,000:", high_salary)
print("Total Salary Expenditure:", total_expenditure)
```

```python
from functools import reduce

marks = [85, 42, 76, 90, 35]

toppers = list(filter(lambda x: x >= 90, marks))
failed = list(filter(lambda x: x < 40, marks))

average = reduce(lambda x, y: x + y, marks) / len(marks)

print("Toppers:", toppers)
print("Failed Students:", failed)
print("Class Average:", average)
```

```python
# 1. Student Result Processing System 
# A college wants to process student marks and identify toppers, failed students, and class average using functional programming.

from functools import reduce

marks = [78, 45, 90, 32, 88, 67, 29]

passed = list(filter(lambda x: x >= 40, marks))
grace_marks = list(map(lambda x: x + 5, marks))
total = reduce(lambda x, y: x + y, marks)
average = total / len(marks)

print("Passed Students Marks:", passed)
print("Grace Marks:", grace_marks)
print("Total Marks:", total)
print("Average Marks:", average)
```

```python
# 2. Online Shopping Discount System Case Study
# An e-commerce website gives:
# • 20% discount for orders above ₹1000
# • Removes orders below ₹100
# • Calculates total revenue
# Program

from functools import reduce

orders = [250, 1200, 560, 50, 3000, 80]

valid_orders = list(filter(lambda x: x >= 100, orders))
discounted = list(map(lambda x: x * 0.8 if x > 1000 else x, valid_orders))
revenue = reduce(lambda x, y: x + y, discounted)

print("Valid Orders:", valid_orders)
print("Discounted Prices:", discounted)
print("Total Revenue:", revenue)
```

```python
# 3. Employee Salary Management

from functools import reduce

salaries = [25000, 55000, 48000, 72000, 35000]

# Salary increment
updated_salary = list(map(lambda x: x * 1.10, salaries))

# High salary employees
high_salary = list(filter(lambda x: x > 50000, updated_salary))

# Total expenditure
total = reduce(lambda x, y: x + y, updated_salary)

print("Updated Salaries:", updated_salary)
print("High Salary Employees:", high_salary)
print("Total Salary:", total)
```

```python
# 4. Hospital Patient Record Analysis
from functools import reduce

temps_c = [36.5, 37.0, 39.2, 38.5, 36.8]

# Convert to Fahrenheit
temps_f = list(map(lambda x: (x * 9 / 5) + 32, temps_c))

# Fever patients
fever = list(filter(lambda x: x > 100, temps_f))

# Average temperature
total = reduce(lambda x, y: x + y, temps_f)
avg = total / len(temps_f)

print("Temperatures in Fahrenheit:", temps_f)
print("Fever Patients:", fever)
print("Average Temperature:", avg)
```

```python
# 5. Electricity Bill Calculation
from functools import reduce

bills = [450, 1200, 800, 300, 1500]

# Add GST
gst_bills = list(map(lambda x: x * 1.18, bills))

# Pending bills
pending = list(filter(lambda x: x > 500, gst_bills))

# Total collection
collection = reduce(lambda x, y: x + y, gst_bills)

print("Bills with GST:", gst_bills)
print("Pending Bills:", pending)
print("Total Collection:", collection)
```

```python
# 6. Food Delivery Rating Analysis Case Study
# A food delivery app wants to:
# • Identify restaurants with rating above 4
# • Increase ratings by 0.2 for premium members
# • Calculate average rating
# Program

from functools import reduce

ratings = [3.5, 4.2, 4.8, 2.9, 4.0]

# Premium bonus
updated = list(map(lambda x: x + 0.2, ratings))

# Top restaurants
top = list(filter(lambda x: x > 4, updated))

# Average rating
total = reduce(lambda x, y: x + y, updated)
avg = total / len(updated)

print("Updated Ratings:", updated)
print("Top Restaurants:", top)
print("Average Rating:", avg)
```
