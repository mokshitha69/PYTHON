```python
# Positivity of a number
a = int(input("Enter a:"))
if a>0:
    print("Number is positive")
else :
    print("Number is negative")
```

```python
# Even or odd
a = int(input("Enter a:"))
if (a / 2==0):
    print("Number is even")
else :
    print("Number is odd")
```

```python
# Voting eligibility
age = int(input("Enter your age"))
if age>=18 :
    print("Eligible to vote")
else :
    print("Not eligible to vote")
```

```python
# Program to find the highest of three numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    print("Highest number is:", a)
elif b >= a and b >= c:
    print("Highest number is:", b)
else:
    print("Highest number is:", c)
```

```python
#printing numbers from 1 to 5 using while loop
i = 1
while i <= 5:
    print(i)
    i = i + 1
```

```python
#printing numbers from 1 to 5 using for loop
for i in range(1, 6):
    print(i)
```

```python
#printing numbers from 1 to 5 skipping at 4
i = 1

while i <= 5:
    if i == 4:
        i += 1
        continue
    print(i)
    i += 1
```

```python
#printing numbers from 1 to 5 stopping at 4
i = 1

while i <= 5:
    if i == 4:
        break
    print(i)
    i += 1
```
