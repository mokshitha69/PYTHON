```python
# Relational operator
a = int(input("enter a:"))
b = int(input("Enter b:"))
print(a<b)
print(a>b)
print(a==b)
print(a<=b)
print(a>=b)
print(a!=b)
```

```python
# Arithmetic operator
a = int(input("enter a:"))
b = int(input("Enter b:"))
print(a + b) 
print(a- b) 
print(a * b)
print(a / b) 
print(a % b) 
print(a ** b) 
print(a // b) 
```

```python
# Bitwise operator
a = int(input("enter a:"))
b = int(input("Enter b:"))
print(a & b) 
print(a | b) 
print(a ^ b) 
```

```python
# Assignment operators
a = int(input("enter a:"))
a+= 45
a-= 7
b = a
print("a=", a)
print("b=", b)
```

```python
# member operators
x = [1, 2, 3]
print(2 in x)
print(5 not in x)
```

```python
# Identity operator
x = [1, 2, 3]
y = [1, 2, 3]
print(x is y)
print(x == y)
```
